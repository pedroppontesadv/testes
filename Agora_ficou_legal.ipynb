```python
import pandas as pd
import plotly.graph_objects as go
from IPython.display import display, HTML

# Carregar o arquivo Excel (Certifique-se de fazer o upload do arquivo para o Colab antes)
df = pd.read_excel('/content/dados_pesquisa_satisfação.xlsx')

# Identificando colunas quantitativas
colunas_quantitativas = [col for col in df.columns if col not in ['Nome', 'Empresa'] and '1 a' in col]

# Gerando gráficos de pizza para perguntas quantitativas
for coluna in colunas_quantitativas:
    media_sem_zeros = df[coluna][df[coluna] != 0].mean()
    valores_zero = len(df[df[coluna] == 0])

    dados_grafico = df[coluna].value_counts().reset_index()
    dados_grafico.columns = ['Resposta', 'Quantidade']
    dados_grafico = dados_grafico[dados_grafico['Resposta'] != 0]

    fig = go.Figure(data=[go.Pie(labels=dados_grafico['Resposta'], values=dados_grafico['Quantidade'], hole=.3)])
    fig.update_layout(title_text=f"{coluna} - Média (sem zeros): {media_sem_zeros:.2f}, Clientes ainda sem contato: {valores_zero}")
    fig.show()

# Dicionário com as perguntas e respostas qualitativas
respostas_qualitativas = {
    "O que o Corplaw significa": [
        "Tulio Maia, Transcol Nordeste: Segurança",
        "Amanda, Vitoria Vidros: Satisfação",
        "Leticia Xavier, Bamboo Construcoes: Equipe qualificada e disponível",
        "Dayana Kelly, Grupo Transcol: Competência"
    ],
    "O que melhorar": [
        "Tulio Maia, Transcol Nordeste: Rapidez",
        "Ademir, Atacadao Rural: Mais rapidez nas ações",
        "Amanda, Vitoria Vidros: Agilidade no processo",
        "Leticia Xavier, Bamboo Construcoes: Melhoria na comunicação"
    ],
    "O que o navegador poderia ser melhor": [
        "Tulio Maia, Transcol Nordeste: Sem comentários",
        "Ademir, Atacadao Rural: Ser mais ágil",
        "Amanda, Vitoria Vidros: Não utilizei",
        "Leticia Xavier, Bamboo Construcoes: Ágil e solícito"
    ],
    "O que o raio-x poderia ser melhor": [
        "Tulio Maia, Transcol Nordeste: Rapidez",
        "Ademir, Atacadao Rural: Simplificar",
        "Amanda, Vitoria Vidros: Não utilizei o raio X jurídico"
    ],
    "O que pode melhorar nos Conselheiros Jurídicos?": [
        "Tulio Maia, Transcol Nordeste: Nada a declarar",
        "Ademir, Atacadao Rural: Maior agilidade",
        "Amanda, Vitoria Vidros: Satisfeito"
    ],
    "Em qual momento você viu valor": [
        "Tulio Maia, Transcol Nordeste: Soluções propostas",
        "Ademir, Atacadao Rural: Na negociação da minha empresa",
        "Amanda, Vitoria Vidros: Iniciando processo junto a Corplaw"
    ],
    "Em qual momento você não gostou": [
        "Tulio Maia, Transcol Nordeste: Não senti",
        "Ademir, Atacadao Rural: Demora em algumas respostas",
        "Leticia Xavier, Bamboo Construcoes: Falta de comunicação"
    ],
    "Quão essencial o Corplaw é para a sua empresa?": [
        "Tulio Maia, Transcol Nordeste: Difícil imaginar sem o Corplaw",
        "Ademir, Atacadao Rural: Útil, porém dispensável",
        "Amanda, Vitoria Vidros: Útil, porém dispensável"
    ],
    "Há alguma dor que você sente que no momento ainda esteja fora do nosso radar?": [
        "Tulio Maia, Transcol Nordeste: Mais créditos para demandas",
        "Ademir, Atacadao Rural: Finalizar processo de negociação"
    ],
    "Há algum serviço novo que você gostaria que o Corplaw prestasse para sua empresa?": [
        "Maioria, Maioria: Vários respondentes expressaram não ter demandas por novos serviços no momento."
    ],
    "Qual seria o tempo de duração ideal, em horas, para cada um desses treinamentos exclusivos?": [
        "Tulio Maia, Transcol Nordeste: Até 4 horas",
        "Ademir, Atacadao Rural: 1 hora"
    ],
    "Se cada um desses treinamentos exclusivos para a sua empresa fosse oferecido ao preço de R$ 9.500,00, você consideraria esse valor:": [
        "Tulio Maia, Transcol Nordeste: Alto",
        "Ademir, Atacadao Rural: Muito Alto",
        "Amanda, Vitoria Vidros: Muito Alto",
        "Leticia Xavier, Bamboo Construcoes: Alto"
    ]
}

# Exibindo as tabelas para as respostas qualitativas
for pergunta, respostas in respostas_qualitativas.items():
    # Criação da estrutura da tabela HTML
    tabela_html = f"<h3>{pergunta}</h3>"
    tabela_html += "<table border='1'>"
    tabela_html += "<tr><th>Responsável</th><th>Empresa</th><th>Resposta</th></tr>"
    for resposta in respostas:
        # Dividindo a resposta para extrair as informações corretamente
        partes_resposta = resposta.split(": ")
        nome_empresa = partes_resposta[0].split(', ')
        nome = nome_empresa[0]  # Nome do responsável
        empresa = nome_empresa[1] if len(nome_empresa) > 1 else ""  # Nome da empresa
        resposta_texto = partes_resposta[1] if len(partes_resposta) > 1 else ""

        # Adicionando as informações na tabela HTML
        tabela_html += f"<tr><td>{nome}</td><td>{empresa}</td><td>{resposta_texto}</td></tr>"

    tabela_html += "</table><br>"
    display(HTML(tabela_html))
```

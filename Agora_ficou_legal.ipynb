```python
import pandas as pd
import plotly.graph_objects as go
from IPython.display import display, HTML

# Carregar o arquivo Excel (Certifique-se de fazer o upload do arquivo para o Colab antes)
df = pd.read_excel('/content/dados_pesquisa_satisfao.xlsx')

# Identificando colunas quantitativas
colunas_quantitativas = [col for col in df.columns if col not in ['Nome', 'Empresa'] and '1 a' in col]

# Gerando grficos de pizza para perguntas quantitativas
for coluna in colunas_quantitativas:
    media_sem_zeros = df[coluna][df[coluna] != 0].mean()
    valores_zero = len(df[df[coluna] == 0])

    dados_grafico = df[coluna].value_counts().reset_index()
    dados_grafico.columns = ['Resposta', 'Quantidade']
    dados_grafico = dados_grafico[dados_grafico['Resposta'] != 0]

    fig = go.Figure(data=[go.Pie(labels=dados_grafico['Resposta'], values=dados_grafico['Quantidade'], hole=.3)])
    fig.update_layout(title_text=f"{coluna} - Mdia (sem zeros): {media_sem_zeros:.2f}, Clientes ainda sem contato: {valores_zero}")
    fig.show()

# Dicionrio com as perguntas e respostas qualitativas
respostas_qualitativas = {
    "O que o Corplaw significa": [
        "Tulio Maia, Transcol Nordeste: Segurana",
        "Amanda, Vitoria Vidros: Satisfao",
        "Leticia Xavier, Bamboo Construcoes: Equipe qualificada e disponvel",
        "Dayana Kelly, Grupo Transcol: Competncia"
    ],
    "O que melhorar": [
        "Tulio Maia, Transcol Nordeste: Rapidez",
        "Ademir, Atacadao Rural: Mais rapidez nas aes",
        "Amanda, Vitoria Vidros: Agilidade no processo",
        "Leticia Xavier, Bamboo Construcoes: Melhoria na comunicao"
    ],
    "O que o navegador poderia ser melhor": [
        "Tulio Maia, Transcol Nordeste: Sem comentrios",
        "Ademir, Atacadao Rural: Ser mais gil",
        "Amanda, Vitoria Vidros: No utilizei",
        "Leticia Xavier, Bamboo Construcoes: gil e solcito"
    ],
    "O que o raio-x poderia ser melhor": [
        "Tulio Maia, Transcol Nordeste: Rapidez",
        "Ademir, Atacadao Rural: Simplificar",
        "Amanda, Vitoria Vidros: No utilizei o raio X jurdico"
    ],
    "O que pode melhorar nos Conselheiros Jurdicos?": [
        "Tulio Maia, Transcol Nordeste: Nada a declarar",
        "Ademir, Atacadao Rural: Maior agilidade",
        "Amanda, Vitoria Vidros: Satisfeito"
    ],
    "Em qual momento voc viu valor": [
        "Tulio Maia, Transcol Nordeste: Solues propostas",
        "Ademir, Atacadao Rural: Na negociao da minha empresa",
        "Amanda, Vitoria Vidros: Iniciando processo junto a Corplaw"
    ],
    "Em qual momento voc no gostou": [
        "Tulio Maia, Transcol Nordeste: No senti",
        "Ademir, Atacadao Rural: Demora em algumas respostas",
        "Leticia Xavier, Bamboo Construcoes: Falta de comunicao"
    ],
    "Quo essencial o Corplaw  para a sua empresa?": [
        "Tulio Maia, Transcol Nordeste: Difcil imaginar sem o Corplaw",
        "Ademir, Atacadao Rural: til, porm dispensvel",
        "Amanda, Vitoria Vidros: til, porm dispensvel"
    ],
    "H alguma dor que voc sente que no momento ainda esteja fora do nosso radar?": [
        "Tulio Maia, Transcol Nordeste: Mais crditos para demandas",
        "Ademir, Atacadao Rural: Finalizar processo de negociao"
    ],
    "H algum servio novo que voc gostaria que o Corplaw prestasse para sua empresa?": [
        "Maioria, Maioria: Vrios respondentes expressaram no ter demandas por novos servios no momento."
    ],
    "Qual seria o tempo de durao ideal, em horas, para cada um desses treinamentos exclusivos?": [
        "Tulio Maia, Transcol Nordeste: At 4 horas",
        "Ademir, Atacadao Rural: 1 hora"
    ],
    "Se cada um desses treinamentos exclusivos para a sua empresa fosse oferecido ao preo de R$ 9.500,00, voc consideraria esse valor:": [
        "Tulio Maia, Transcol Nordeste: Alto",
        "Ademir, Atacadao Rural: Muito Alto",
        "Amanda, Vitoria Vidros: Muito Alto",
        "Leticia Xavier, Bamboo Construcoes: Alto"
    ]
}

# Exibindo as tabelas para as respostas qualitativas
for pergunta, respostas in respostas_qualitativas.items():
    # Criao da estrutura da tabela HTML
    tabela_html = f"<h3>{pergunta}</h3>"
    tabela_html += "<table border='1'>"
    tabela_html += "<tr><th>Responsvel</th><th>Empresa</th><th>Resposta</th></tr>"
    for resposta in respostas:
        # Dividindo a resposta para extrair as informaes corretamente
        partes_resposta = resposta.split(": ")
        nome_empresa = partes_resposta[0].split(', ')
        nome = nome_empresa[0]  # Nome do responsvel
        empresa = nome_empresa[1] if len(nome_empresa) > 1 else ""  # Nome da empresa
        resposta_texto = partes_resposta[1] if len(partes_resposta) > 1 else ""

        # Adicionando as informaes na tabela HTML
        tabela_html += f"<tr><td>{nome}</td><td>{empresa}</td><td>{resposta_texto}</td></tr>"

    tabela_html += "</table><br>"
    display(HTML(tabela_html))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from wordcloud import WordCloud
from textblob import TextBlob
from nltk.corpus import stopwords
from collections import Counter

df = pd.read_excel('/content/dados_pesquisa_satisfa√ß√£o.xlsx')

# Fun√ß√£o para an√°lise quantitativa
def analise_quantitativa(coluna):
    media = df[coluna].mean()
    plt.figure(figsize=(10, 6))
    df[coluna].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.axhline(y=media, color='r', linestyle='-', linewidth=2, label=f'M√©dia: {media:.2f}')
    plt.title(f'An√°lise Quantitativa de {coluna}')
    plt.xlabel('Valores')
    plt.ylabel('Frequ√™ncia')
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Fun√ß√£o para gerar wordcloud a partir de respostas textuais
def gerar_wordcloud(coluna):
    stop_words = set(stopwords.words('portuguese'))
    stop_words.update(['de', 'a', 'o', 'que', 'e', 'do', 'da', 'em', 'um', 'para', 'com', 'nada', 'nao', 'grupo', 'alguns', 'sempre', 'corplaw', 'n√£o', 'uma', 'os', 'no', 'tudo', 'ainda', 'algumas', 'x', 'bem', 'se', 'na', 'por', 'mais', 'as', 'dos', 'como', 'mas', 'ao', 'ele', 'das', '√†', 'seu', 'sua', 'ou'])
    texto = ' '.join(df[coluna].dropna()).lower()
    palavras = [palavra for palavra in texto.split() if palavra not in stop_words]
    contagem_palavras = Counter(palavras)
    palavras_mais_citadas = contagem_palavras.most_common(5)
    wordcloud_input = ' '.join([palavra for palavra, _ in palavras_mais_citadas])
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(wordcloud_input)
    plt.figure(figsize=(10, 7))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Fun√ß√£o para an√°lise de sentimentos (aproxima√ß√£o simplificada)
def analise_sentimentos(coluna):
    sentimentos = df[coluna].dropna().apply(lambda x: TextBlob(x).sentiment.polarity)
    media_sentimentos = sentimentos.mean()
    if media_sentimentos > 0:
        return "Somos extremamente necess√°rios"
    elif media_sentimentos < 0:
        return "N√£o somos necess√°rios"
    else:
        return "Somos necess√°rios"

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 1 a 10 quanto nossos clientes recomendam o Corplaw :")
print(analise_quantitativa('1 a 10 quanto recomenda o Corplaw'))

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 1 a 5 quanto nossos clientes est√£o satisfeito com as demandas : ")
print(analise_quantitativa('1 a 5 quanto est√° satisfeito com as demandas'))

# Gerando um wordcloud para respostas textuais
# Substitua 'Comentarios_Gerais' pelo nome real da coluna de coment√°rios
print("Como podemos melhorar nossas entregas :")
gerar_wordcloud('o que melhorar')

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 1 a 5 quanto nossos clientes est√£o satisfeito com o Navegador : ")
print(analise_quantitativa('1 a 5 quanto est√° satisfeito com o navegador'))

# Gerando um wordcloud para respostas textuais
# Substitua 'Comentarios_Gerais' pelo nome real da coluna de coment√°rios
print("Como nossos clientes veem nossos Navegadores :")
gerar_wordcloud('o que o navegador poderia ser melhor')

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 1 a 5 quanto nossos clientes est√£o satisfeito com o Raio-X : ")
print(analise_quantitativa('1 a 5 quanto est√° a satisfa√ß√£o com o raio-x'))

# Gerando um wordcloud para respostas textuais
# Substitua 'Comentarios_Gerais' pelo nome real da coluna de coment√°rios
print("Como vamos melhorar a nossa entrega do Raio-X :")
gerar_wordcloud('o que o raio-x poderia ser melhor')

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 1 a 5 quanto nossos clientes est√£o satisfeito com o Conselheiro Jur√≠dico : ")
print(analise_quantitativa('1 a 5 quanto est√° a satisfa√ß√£o com o conselheiro jur√≠dico'))

# Gerando um wordcloud para respostas textuais
# Substitua 'Comentarios_Gerais' pelo nome real da coluna de coment√°rios
print("Como nossos clientes veem nosso servi√ßo de Conselheiros Jur√≠dicos:")
gerar_wordcloud('O que pode melhorar nos Conselheiros Jur√≠dicos?')

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 1 a 5, qual o grau de satisfa√ß√£o com o Time Consultivo : ")
print(analise_quantitativa('De 1 a 5, qual o seu grau de satisfa√ß√£o com o Time Consultivo?'))

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 1 a 5, qual seu grau de satisfa√ß√£o com o Time Trabalhista : ")
print(analise_quantitativa('De 1 a 5, qual o seu grau de satisfa√ß√£o com o Time Trabalhista?'))

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 1 a 5, qual o grau de satisfa√ß√£o com o Time Societ√°rio : ")
print(analise_quantitativa('De 1 a 5, qual o seu grau de satisfa√ß√£o com o Time Societ√°rio?'))

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 1 a 5, qual o grau de satisfa√ß√£o com o Time Tribut√°rio? : ")
print(analise_quantitativa('De 1 a 5, qual o seu grau de satisfa√ß√£o com o Time Tribut√°rio?'))

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 1 a 5, qual o grau de satisfa√ß√£o com o Time de Propriedade Intelectual : ")
print(analise_quantitativa('De 1 a 5, qual o seu grau de satisfa√ß√£o com o Time de Propriedade Intelectual?'))

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 1 a 5, qual o seu grau de satisfa√ß√£o com o Time Contencioso Judicial : ")
print(analise_quantitativa('De 1 a 5, qual o seu grau de satisfa√ß√£o com o Time Contencioso Judicial?'))

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 01 a 05, quanto nosso clientes interessariam por uma sequ√™ncia de Masterclasses em grupo com outras empresas na √°rea trabalhista : ")
print(analise_quantitativa('De 01 a 05, quanto voc√™ se interessaria por uma sequ√™ncia de Masterclasses em grupo com outras empresas na √°rea trabalhista?'))

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 01 a 05, quanto nosso clientes interessariam por uma sequ√™ncia de Masterclasses em grupo com outras empresas na √°rea societ√°ria : ")
print(analise_quantitativa('De 01 a 05, quanto voc√™ se interessaria por uma sequ√™ncia de Masterclasses em grupo com outras empresas na √°rea societ√°ria?'))

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 01 a 05, quanto nosso clientes interessariam por uma sequ√™ncia de Masterclasses em grupo com outras empresas na √°rea de contratos : ")
print(analise_quantitativa('De 01 a 05, quanto voc√™ se interessaria por uma sequ√™ncia de Masterclasses em grupo com outras empresas na √°rea de contratos?'))

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 01 a 05, quanto nosso clientes interessariam por um treinamento exclusivo na √°rea trabalhista : ")
print(analise_quantitativa('De 01 a 05, quanto voc√™ se interessaria por um treinamento online exclusivo para a sua empresa na √°rea trabalhista?'))

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 01 a 05, quanto nosso clientes interessariam por um treinamento exclusivo na √°rea societ√°ria : ")
print(analise_quantitativa('De 01 a 05, quanto voc√™ se interessaria por um treinamento online exclusivo para a sua empresa na √°rea societ√°ria?'))

# Realizando a an√°lise quantitativa
# Substitua 'Nota_Conselheiros_Jur√≠dicos' pelo nome real da coluna
print("De 01 a 05, quanto nosso clientes interessariam por um treinamento exclusivo na √°rea de contratos : ")
print(analise_quantitativa('De 01 a 05, quanto voc√™ se interessaria por um treinamento online exclusivo para a sua empresa na √°rea de contratos?'))

# Gerando um wordcloud para respostas textuais
# Substitua 'Comentarios_Gerais' pelo nome real da coluna de coment√°rios
print("O que o Corplaw significa : ")
gerar_wordcloud('O que o Corplaw significa')

# Gerando um wordcloud para respostas textuais
# Substitua 'Comentarios_Gerais' pelo nome real da coluna de coment√°rios
print("Quando nossos clientes mais viram valor no Corplaw : ")
gerar_wordcloud('em qual momento voc√™ viu valor')

# Gerando um wordcloud para respostas textuais
# Substitua 'Comentarios_Gerais' pelo nome real da coluna de coment√°rios
print("Quando nossos cliente n√£o gostaram da entrega : ")
gerar_wordcloud('em qual momento voc√™ n√£o gostou')

# Analisando os sentimentos das respostas subjetivas
# Substitua 'Comentarios_Gerais' pelo nome real da coluna de coment√°rios
print("Qu√£o essencial o Corplaw √© para nossos clientes?")
print(analise_sentimentos('Qu√£o essencial o Corplaw √© para a sua empresa?'))


```
